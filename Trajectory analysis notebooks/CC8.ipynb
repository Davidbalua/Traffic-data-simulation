```python
from datetime import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

scenario 5 is based scenario generated by me: all are 3.5


```python
#scenario 5 is based scenario generated by me: all are 3.5
vehicle_routes_scenario5 = pd.read_csv("data/Scenarios generated by me/CC8/GothaerPlatz_scenario5_allare3Point5.csv")
#change the vehicle type from number to name
vehicle_routes_scenario5['VEHTYPE'].replace(to_replace=[100, 150, 200, 300], value=["Car", "Transporter", "HGV", "Bus"],
                                            inplace=True)
vehicle_routes_filtered_scenario5 = vehicle_routes_scenario5[vehicle_routes_scenario5["$VEHICLE:NO"]<318]
max_accs_cars_scenario5= vehicle_routes_filtered_scenario5.groupby(["$VEHICLE:NO"]).max()
```

```python
max_accs_cars_scenario5
```

```python
# barchart with dictionary
vehicle_numbers = [i for i in range(1, 318)]
fig, ax = plt.subplots(figsize=(10, 10))
ax.bar(vehicle_numbers, max_accs_cars_scenario5["ACCELERATION"], label="Vehicle")
ax.set_title(
    label="Maximum acceleration of all vehicles from standstill position in the based scenario with the default CC8",
    color="red")
# draw CC8
x_coordinates = [0, 318]

y_coordinates = [3.5, 3.5]

ax.plot(x_coordinates, y_coordinates, color="red", label="CC8 Parameter")
ax.set_xlabel("Vehicle numbers", color="red")
ax.set_ylabel("Maximum Acceleration(m/s2)", color="red")
ax.tick_params(axis="both", length=15, direction="inout", color="blue", labelcolor="red")
ax.spines["bottom"].set_color("red")
ax.spines["left"].set_color("red")
# draw legends
ax.legend(loc="lower right", shadow="true", fancybox=True, title="Vehicle Number");
ax.grid("on", alpha=0.9)

# plt.savefig("data/CC8/Based_Scenario_by_me.png")
max_accs_cars_scenario5["ACCELERATION"].describe()
```

scenario 3 cc8 = 10 and max and desired functions are 10


```python
#scenario 3 cc8 = 10 and max and desired functions are 10
#change the vehicle type from number to name
vehicle_routes_scenario3 = pd.read_csv("data/Scenarios generated by me/CC8/GothaerPlatz_0_to_500_seconds_cc8_10.csv")
#change the vehicle type from number to name
vehicle_routes_scenario3['VEHTYPE'].replace(to_replace=[100, 150, 200, 300], value=["Car", "Transporter", "HGV", "Bus"],
                                            inplace=True)
max_accs_cars_scenario3 = vehicle_routes_scenario3.groupby(["$VEHICLE:NO"]).max()

fig, ax = plt.subplots(figsize=(10, 10))
ax.bar(vehicle_numbers, max_accs_cars_scenario3["ACCELERATION"], label="Vehicle")
ax.set_xlabel("Vehicle numbers", color="red")
ax.set_ylabel("Maximum Acceleration(m/s2)", color="red")
ax.set_title(
    label="Maximum acceleration of all vehicles from standstill position in the scenario with CC8 = 10 m/s2 and the maximum value of desired and maximum acceleration functions are 10 m/s2",
    color="red", fontsize = 8)
# draw default CC8
x_coordinates = [0, 318]
y_coordinates = [3.5, 3.5]

# draw CC8 10
x1_coordinates = [0, 318]
y1_coordinates = [10, 10]

ax.plot(x_coordinates, y_coordinates, color="red", label="Default CC8 Parameter")
ax.plot(x1_coordinates, y1_coordinates, color="green", label="Updated CC8 Parameter")
ax.tick_params(axis="both", length=15, direction="inout", color="blue", labelcolor="red")
ax.spines["bottom"].set_color("red")
ax.spines["left"].set_color("red")

# draw legends
ax.legend(loc="lower right", shadow="true", fancybox=True, title="Vehicle Number");
ax.grid("on", alpha=0.9)
# plt.savefig("data/CC8/cc8 10 and max and desired functions are 10.png")
max_accs_cars_scenario3["ACCELERATION"].describe()
```

calculating normal distribution



```python
#https://stackoverflow.com/questions/10138085/how-to-plot-normal-distribution
import scipy.stats as stats
import math

fig, ax = plt.subplots(figsize=(10, 10))
mu_baseline = 1.829
sigma_baseline = 0.947
x_baseline = np.linspace(mu_baseline - 3*sigma_baseline, mu_baseline + 3*sigma_baseline, 100)
line_baseline, =ax.plot(x_baseline, stats.norm.pdf(x_baseline, mu_baseline, sigma_baseline), label="Baseline Scenario")
line_baseline.set_color("red")
ax.legend(loc="lower right", shadow="true", fancybox=True, title="Name of Scenario");

mu_scenario1 = 2.030
sigma_scenario1 = 1.217
x_scenario1 = np.linspace(mu_scenario1 - 3*sigma_scenario1, mu_scenario1 + 3*sigma_scenario1, 100)
line_scenario1, =ax.plot(x_scenario1, stats.norm.pdf(x_scenario1, mu_scenario1, sigma_scenario1), label="Scenario 1")
line_scenario1.set_linestyle("--")
ax.xaxis.grid(True, color="black", linewidth=0.2)
ax.legend(loc="lower right", shadow="true", fancybox=True, title="Name of Scenario");



mu_scenario2 = 1.558
sigma_scenario2 = 0.698
x_scenario2 = np.linspace(mu_scenario2 - 3*sigma_scenario2, mu_scenario2 + 3*sigma_scenario2, 100)
line_scenario2, =ax.plot(x_scenario2, stats.norm.pdf(x_scenario2, mu_scenario2, sigma_scenario2), label="Scenario 2")
line_scenario2.set_linestyle("--")
line_scenario2.set_color("blue")
ax.xaxis.grid(True, color="black", linewidth=0.2)
ax.legend(loc="lower right", shadow="true", fancybox=True, title="Name of Scenario");



plt.xticks(fontsize=10, color="red")
plt.yticks(fontsize=10, color="red")
ax.set_xlabel("Acceleration", fontsize=10, color="red", rotation="45")
ax.set_title(label="The normal distribution for the maximum standstill acceleration of all vehicles in three scenarios",color="red")
# plt.savefig("data/CC8/distributions.png")

```
